#Working with dictionaries

Adding key-value pair to dictionary

```python
student = {"college": "DJSCE", "marks": "88%"}
student["name"] = "Mayur"
print(student)

student["marks"]="85%"
print(student)

del student["name"]
print(student)
```

Looping in dictionaries

```python
foods = {"Indian": "Dal Rice", "Italian": "Pizza", "Chinese": "Noodles"}

#looping through items of dictionaries
#    key   value
for cusine, food in foods.items():
  print("Cusine: "+cusine+"\n"+"Favourite food"+food)
  print("\n")

#looping through keys of dictionaries
for cusine in foods.keys():
  print(cusine)

print("\n")
#looping through values of dictionaries
for food in foods.values():
  print(food)

if "American" not in foods.values():
  print("\nPlease enter a American cusine food too\n")

foods["American"] = "Burger"
print(foods)
```

6-6. Polling: Use the code in favorite_languages.py (page 104).

• Make a list of people who should take the favorite languages poll. Include
some names that are already in the dictionary and some that are not.

• Loop through the list of people who should take the poll. If they have
already taken the poll, print a message thanking them for responding.
If they have not yet taken the poll, print a message inviting them to take
the poll.

```python
favorite_languages = {
  'jen': 'python',
  'sarah': 'c',
  'edward': 'ruby',
  'phil': 'python',
  }

people = ['jen', 'mayur', 'phil', 'eric', 'sarah']

for person in people:
  if person not in favorite_languages.keys():
    print("Hey, "+person+" please take our poll")
  else :
    print("Thank you "+person+" for taking our poll")
```

6-5. Rivers: Make a dictionary containing three major rivers and the country
each river runs through. One key-value pair might be 'nile': 'egypt'.

• Use a loop to print a sentence about each river, such as The Nile runs
through Egypt.

• Use a loop to print the name of each river included in the dictionary.

• Use a loop to print the name of each country included in the dictionary.

```python
rivers = {"Ganga": "India", "Amazon River": "Brazil", "Nile": "Egypt"}

for river, country in rivers.items():
  print("The "+river+" runs through "+country)

print("\nRiver included in the dictionary:\n")
for river in rivers.keys():
  print(river)

print("\nCountry included in the dictionary:\n")
for country in rivers.values():
  print(country)
```
