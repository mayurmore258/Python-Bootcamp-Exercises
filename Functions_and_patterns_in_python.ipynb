```python
def even_numbers(a, b):
  print(f"Even numbers form {a} to {b} are:")
  for value in range(a,b+1):
    if value % 2 == 0:
      print(value)

even_numbers(10,30)
```

```python
#pattern printing

def square(n):
  for i in range(n):
    for j in range(n):
      print("*", end=" ")
    print(" ")
square(5)

print()
def right_trinagle(n):
  for i in range(n):
    for j in range(i+1):
      print("*", end=" ")
    print(" ")
right_trinagle(5)

print()
def upsidedown_right_tringle(n):
  for i in range(n):
    for j in range(n-i):
      print("*", end=" ")
    print(" ")
upsidedown_right_tringle(5)

print()
def flipped_right_triangle(n):
  for i in range(1,n+1):
    for j in range(n-i):
      print(" ", end=" ")
    for j in range(i):
      print("*", end=" ")
    print()
flipped_right_triangle(5)

print()
def flipped_upsidedown_right_triangle(n):
  for i in range(n):
    for j in range(i):
      print(" ", end=" ")
    for j in range(n-i):
      print("*", end=" ")
    print(" ")
flipped_upsidedown_right_triangle(5)

print()
def mountain(n):
  for i in range(1,n+1):
    for j in range(n-i):
      print(" ", end=" ")
    for j in range(i):
      print("*", end=" ")
    for j in range(i-1):
      print("*", end=" ")
    print()
mountain(5)

print()
def upsidedown_mountain(n):
  for i in range(n):
    for j in range(i):
      print(" ", end=" ")
    for j in range(n-i):
      print("*", end=" ")
    for j in range(n-i-1):
      print("*", end=" ")
    print()
upsidedown_mountain(5)

print()
def diamond(n):
  for i in range(n):
    for j in range(n-i):
      print(" ", end=" ")
    for j in range(i):
      print("*", end=" ")
    for j in range(i-1):
      print("*", end=" ")
    print()
  for i in range(n):
    for j in range(i):
      print(" ", end=" ")
    for j in range(n-i):
      print("*", end=" ")
    for j in range(n-i-1):
      print("*", end=" ")
    print()
diamond(5)
```

```python
#Function with return type

def check_even(a,b):
  if a%2==0 and b%2==0:
    return "Both are even";
  elif a%2==0:
    return a;
  elif b%2==0:
    return b;
  else:
    return "No result";

even = check_even(5,9)
print(even)
```
